```
import pandas as pd

df = pd.read_excel('/content/Compras2022.xlsx')
display(df.head())
```

### Análise Inicial do DataFrame

Vamos começar examinando as informações gerais do DataFrame para entender os tipos de dados e a presença de valores não nulos.

```
df.info()
```

### Estatísticas Descritivas

Agora, vamos olhar as estatísticas descritivas para colunas numéricas, o que pode revelar anomalias ou distribuições incomuns.

```
display(df.describe())
```

### Verificação de Valores Ausentes

É crucial identificar a quantidade de valores ausentes em cada coluna para decidir as estratégias de tratamento (remoção, preenchimento, etc.).

```
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage (%)': missing_percentage})
display(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))
```

### Verificação de Valores Únicos para Colunas Categóricas

Vamos verificar a cardinalidade de algumas colunas que parecem ser categóricas para identificar possíveis inconsistências ou a necessidade de padronização.

```
print('Unique values for proveedor:')
print(df['proveedor'].value_counts().head())
print('\nUnique values for producto:')
print(df['producto'].value_counts().head())
```

```
# Limpeza da descrição
df['descripcion'] = (
    df['descripcion']
        .astype(str)
        .str.upper()
        .str.normalize('NFKD')
        .str.encode('ascii', errors='ignore')
        .str.decode('utf-8')
        .str.strip()
)
```

```
# Remover duplicados
df = df.drop_duplicates()
df.head()
```

```
# Exportar versão limpa
df.to_excel("Compras2022_clean.xlsx", index=False)
```
