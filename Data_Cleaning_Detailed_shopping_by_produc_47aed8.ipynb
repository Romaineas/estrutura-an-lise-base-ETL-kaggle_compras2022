#### Neste caso, temos o mesmo valor nos dados Nan e null, então vamos descartar os dados Nan juntamente com os dados null.



````python
# Drop data Nan
df.dropna(inplace = True)
````

### Nas informações de dados, vemos o tipo de cada dado; portanto, verificaremos o número de dados únicos do tipo objeto.

````python
df.columns.unique()
````

````python
df = df.rename(columns = {'proveedor' : 'Supplier name', 'fecha' : 'Date of purchase', 'tipo' : 'type', 'numero' : 'number', 'bodega' : 'store', 'producto' : 'product',
       'descripcion' : 'description', 'cant' : 'amount', 'unidad' : 'unit', 'vlrunit' : 'vlr unit', 'total_siniva' : 'total without VAT', 'iva_tasa' : 'vat_rate',
       'valor_iva' : 'vat_value', 'cuenta' : 'account', 'precio_actual' : 'Actual Price', 'precio_nuevo': 'new_price', 'saldo_actual' : 'current balance',
       'margen_bruto' : 'gross margin'})

df
````

````python
Supplier_name = df['Supplier name'].nunique()
type          = df['type'].nunique()
store         = df['store'].nunique()
product       = df['product'].nunique()
description   = df['description'].nunique()
unit          = df['unit'].nunique()
account       = df['account'].nunique()
````

````python
print(f' Supplier_name : {Supplier_name} \n',
      f'type          : {type}\n',
      f'store         : {store}\n',
      f'product       : {product}\n',
      f'description   : {description}\n',
      f'unit          : {unit}\n',
      f'account       : {account}\n')
````

#### For the small number unique, we will transform it to integer

### Check the data unique each column

````python
type          = df['type'].unique()
store         = df['store'].unique()
account       = df['account'].unique()
````

````python
print(f' type          : {type}\n',
      f'store         : {store}\n',
      f'account       : {account}\n')
````

````python
# Transform data to integer
df['type'] = df['type'].replace(['FC', 'MR', 'RM'], ['1', '2', '3'])
df['account'] = df['account'].replace([14351601, 14350101, 14350501, 14353001, 14353010], [1, 2, 3, 4, 5])
````



```
# Isto está formatado como código
```

#### Podemos ver que a coluna `product` contém dados em `int`, mas o tipo de dados é `object`, então vamos alterá-lo para `int64`.

But we have the problem in change the type of data, in column product we have the data not in integer instead some data in string object so we will drop all the string data

````python
# Change all data string to Nan and drop it
df['product'] = df['product'].apply(lambda x: np.nan if isinstance(x, str) else x).dropna()
df.dropna(inplace = True)
````

````python
# Change the data type
df['type'] = df['type'].astype('int64')
df['account'] = df['account'].astype('int64')
df['store'] = df['store'].astype('int64')
df['product'] = df['product'].astype('int64')
````

Na tabela temos uma coluna com 'preço atual' e uma coluna com 'novo preço'.
Então, vou criar uma nova coluna com 'Preço_Atual'.

````python
df.insert(15, 'Actual_Price', df.apply(lambda row: row['new_price'] if row['new_price'] != 0 else row['Actual Price'], axis=1))
df.drop(['Actual Price', 'new_price'], axis = 1, inplace = True)

df.insert(13, 'Total with VAT', df.apply(lambda x: x['total without VAT'] if x['vat_value'] == 0 else x['total without VAT'] - x['vat_value'], axis=1))
````

````python
# Save the change
df.to_csv('/kaggle/working/shopping-by-product_Clean.csv',index=False)
````
